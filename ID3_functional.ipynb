```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("play_tennis.csv")
df = df.sample(frac=1) # (shuffle) modelin öğrenme sürecinde veri sırasının etkisini azaltmak için kullanılır. frac=1, veri setinin tamamını karıştıracağını belirtir.
df
```

```python
X = pd.get_dummies(df.drop('Play Tennis', axis=1))  #get dummies ile hedef özellik dışında kullanacağımız özellikleri sayısal değerlere çeviriyoruz, bunun için numpy array de kullanılabilir ama her özellik için manuel ayarlama yapmak gerekir
y = df['Play Tennis']
X
```

```python
def calculate_entropy(y):
    counts = 0
    counts = y.value_counts()  #Play Tennis'te kaç yes-no olduğunu hesaplar
    #print(counts)
    entropy = 0
    for count in counts:    #counts={9,5}
        probability = count / len(y)
        entropy -= probability * math.log2(probability)
        #print(f"count:", {count})
        #print(f"probability:", probability)
        #print(f"entropy:", entropy)

    return entropy
```

```python
calculate_entropy(y)
```

```python
def calculate_information_gain(X, y, feature):
    total_entropy = 0
    values = 0
    total_entropy = calculate_entropy(y)
    values = X[feature].unique()
    #print(f"values:", values)
    weighted_entropy = 0
    for value in values:
        subset_y = y[X[feature] == value]
        #print(f"\nvalue:", value)
        #print(f"subset:\n", subset_y)
        weighted_entropy += len(subset_y) / len(y) * calculate_entropy(subset_y)
    return total_entropy - weighted_entropy
```

```python
X = df.drop('Play Tennis', axis=1)  # get_dummies uygulanmadı
calculate_information_gain(X, y, 'Outlook')
```

```python
def find_best_feature(X, y):
    best_feature = None
    best_gain = 0
    for feature in X.columns:
        #print(f"\n\nfeature:", feature)
        gain = calculate_information_gain(X, y, feature)
        #print(f"gain:", gain)
        if gain > best_gain:
            best_gain = gain
            best_feature = feature
            #print(f"best_gain:", best_gain)
            #print(f"best_feature:", best_feature)
    return best_feature
```

```python
find_best_feature(X, y)
```

```python
def id3(X, y):
    if len(y.unique()) == 1:  #y tek unique değer içeriyorsa, hepsi yes veya no ise dallanma yapmaya gerek yok
        return y.iloc[0]

    if X.empty:
        return y.mode()[0]

    best_feature = find_best_feature(X, y)
    print(f"first best feature:", best_feature)
    tree = {best_feature: {}}  #dictionary yapısında boş tree oluştu

    for value in X[best_feature].unique():
        subset_X = X[X[best_feature] == value].drop(columns=[best_feature])
        subset_y = y[X[best_feature] == value]
        subtree = id3(subset_X, subset_y)
        tree[best_feature][value] = subtree
        print(f"\n\nbest_feature:", best_feature)
        print(f"value:", value)
        print(f"subset_X:\n", subset_X)
        print(f"subset_y:\n", subset_y)
        print(f"subtree:\n", subtree)
    return tree
```

```python
tree = id3(X, y)
print(tree)
```

```python
def predict(tree, sample):
    if not isinstance(tree, dict):  #leaf node ise değerini döndür (yes-no)
        return tree
    feature = next(iter(tree))
    value = sample[feature]
    subtree = tree[feature].get(value, None)
    if subtree is None:
        return None
    return predict(subtree, sample)

y_pred = X.apply(lambda x: predict(tree, x), axis=1)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

accuracy = accuracy_score(y, y_pred)
cm = confusion_matrix(y, y_pred)
precision = precision_score(y, y_pred, pos_label="Yes")  # "Yes" labelini pozitif sınıf olarak alıyoruz
recall = recall_score(y, y_pred, pos_label="Yes")
f1 = f1_score(y, y_pred, pos_label="Yes")

print(f"Sklearn Accuracy: {accuracy}")  #Accuracy: Doğru tahmin edilen örneklerin toplam örneğe oranıdır
print("Sklearn Confusion Matrix:\n", cm) #TP FP FN TN
print(f"Sklearn Precision: {precision}") #Pozitif olarak tahmin edilen örneklerin ne kadarının gerçekten pozitif olduğunu gösterir
print(f"Sklearn Recall: {recall}")  #Recall, true positive örneklerin ne kadarının doğru tahmin edildiğini gösterir
print(f"Sklearn F1-Score: {f1}") #Precision ve Recall'un harmonik ortalaması
```

```python
def custom_accuracy_score(y, y_pred):
    correct = 0
    for true, pred in zip(y, y_pred):  #her iki listenin aynı indexlerini birleştirir, böylelikle (true, pred) şeklinde yeni bir liste oluşturmuş oluruz
        if true == pred:
            correct += 1
    accuracy = correct / len(y)
    return accuracy

custom_accuracy = custom_accuracy_score(y.tolist(), y_pred.tolist())
print(f"Custom Accuracy: {custom_accuracy}")
```

```python
def custom_confusion_matrix(y, y_pred):
    labels = ['Yes', 'No']
    matrix = {label: {label2: 0 for label2 in labels} for label in labels}

    print("Initial Confusion Matrix (All zeros):")
    for true_label, preds in matrix.items():
        print(f"{true_label}: {preds}")

    for true, pred in zip(y, y_pred):
        matrix[true][pred] += 1

    return matrix

custom_cm = custom_confusion_matrix(y.tolist(), y_pred.tolist())
print("Custom Confusion Matrix:")
for true_label, preds in custom_cm.items():
    print(f"{true_label}: {preds}")
```

```python
def custom_precision(y, y_pred, pos_label):
    TP = FP = 0
    for true, pred in zip(y, y_pred):
        if true == pos_label and pred == pos_label:
            TP += 1  # True Positive
        elif true != pos_label and pred == pos_label:
            FP += 1  # False Positive
    return TP / (TP + FP) if (TP + FP) > 0 else 0

def custom_recall(y, y_pred, pos_label="Yes"):
    TP = FN = 0
    for true, pred in zip(y, y_pred):
        if true == pos_label and pred == pos_label:
            TP += 1  # True Positive
        elif true == pos_label and pred != pos_label:
            FN += 1  # False Negative
    return TP / (TP + FN) if (TP + FN) > 0 else 0

def custom_f1(y, y_pred, pos_label="Yes"):
    precision = custom_precision(y, y_pred, pos_label)
    recall = custom_recall(y, y_pred, pos_label)
    return 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

custom_precision_value = custom_precision(y.tolist(), y_pred.tolist(), "Yes")
custom_recall_value = custom_recall(y.tolist(), y_pred.tolist(), "Yes")
custom_f1_value = custom_f1(y.tolist(), y_pred.tolist(), "Yes")

print(f"Custom Precision: {custom_precision_value}")
print(f"Custom Recall: {custom_recall_value}")
print(f"Custom F1-Score: {custom_f1_value}")
```
