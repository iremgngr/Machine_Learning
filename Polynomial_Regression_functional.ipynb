```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

df = pd.read_csv("student.csv")
df.head()
```

```python
y = df["Marks"].values                      # Converts to numpy array format
X = df["time_study"].values.reshape(-1, 1)  # X'i 2D'ye dönüştürüyor
```

```python
def polynomial_features(X, degree):

    features_list = []

    for i in range(degree + 1):  #0'dan degree'ye kadar X'in üslerini bul
        feature = X ** i
        features_list.append(feature)

    # Features list'i yatay olarak birleştirerek bir numpy dizisi oluşturuyor, böylelikle matris haline getiriyoruz
    poly_features = np.hstack(features_list)
    print("Original X:\n", X)
    print("Polynomial Features (degree={}):\n".format(degree), poly_features)
    return poly_features
```

```python
def gradient_descent(X_poly, y, learning_rate=0.00001, iterations=1000):
    m, n = X_poly.shape
    theta = np.random.randn(n)  # Katsayıları rastgele belirle

    for _ in range(iterations):
        ypred = X_poly @ theta  #matris çarpımı
        errors = ypred - y
        gradient = (2 / m) * (X_poly.T @ errors)  #mse'nin türevini bulmak için transpose ile matris çarpımı yap
        theta -= learning_rate * gradient

    return theta
```

```python
def fit_polynomial_regression(X, y, degree, learning_rate, iterations):
    X_poly = polynomial_features(X, degree)
    theta = gradient_descent(X_poly, y, learning_rate, iterations)
    return theta
```

```python
def predict_poly(X, theta, degree):
    X_poly = polynomial_features(X, degree)
    return X_poly @ theta
```

```python
degree = 2
theta = fit_polynomial_regression(X, y, degree, learning_rate=0.001, iterations=1000)

print("Optimized theta:", theta)
```

```python
import matplotlib.pyplot as plt

X_new = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)  # min ve max değerleri arasında 100 eşit aralıklı nokta oluştur
ypred_new = predict_poly(X_new, theta, degree)

plt.scatter(X, y, color='blue', label='Data', s=15)
plt.plot(X_new, ypred_new, color='red', label='Polynomial Regression')
plt.xlabel('Time Study')
plt.ylabel('Marks')
plt.legend()
plt.show()
```
